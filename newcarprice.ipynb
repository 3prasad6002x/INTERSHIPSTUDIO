```python
Q1 = new_price.quantile(q=.25)
Q3 = new_price.quantile(q=.75)
IQR = Q3 - Q1
IQR
```

```python
new_price = new_price[~((new_price < (Q1 - 1.5 * IQR)) |(new_price > (Q3 + 1.5 * IQR))).any(axis=1)]
new_price.shape
```

```python
new_price=new_price.drop(['Year'],axis=1)
new_price.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le_make  = LabelEncoder()
le_model = LabelEncoder()
le_transmission = LabelEncoder()
le_drive_mode = LabelEncoder()
le_vehicle_size = LabelEncoder()
le_vehicle_style = LabelEncoder()
```

```python
new_price=new_price.rename(columns={'Drive Mode':'Drive_Mode','Vehicle Size':'Vehicle_Size','Vehicle Style':'Vehicle_Style'})
```

```python
le_make.fit(new_price.Make)
le_model.fit(new_price.Model)
le_transmission.fit(new_price.Transmission)
le_drive_mode.fit(new_price.Drive_Mode)
le_vehicle_size.fit(new_price.Vehicle_Size)
le_vehicle_style.fit(new_price.Vehicle_Style)
```

```python
new_price['Make']=le_make.transform(new_price['Make'])
new_price['Model']=le_model.transform(new_price['Model'])
new_price['Transmission']=le_transmission.transform(new_price['Transmission'])
new_price['Drive_Mode']=le_drive_mode.transform(new_price['Drive_Mode'])
new_price['Vehicle_Size']=le_vehicle_size.transform(new_price['Vehicle_Size'])
new_price['Vehicle_Style']=le_vehicle_style.transform(new_price['Vehicle_Style'])
new_price
```

```python
Y_train_data = new_price.iloc[:, -2].values

# Independent Variables
X_train_data = new_price.iloc[:,0 : -1].values

# Independent Variables for test Set
X_test = new_price.iloc[:,:].values
```

```python
Y_train_data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, Y_train, Y_val = train_test_split(X_train_data, Y_train_data, test_size = 0.2, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train
```

```python
#Scaling Original Training Data
X_train_data = sc.fit_transform(X_train_data)
print(X_train_data)
#print("--"*20)
#Reshaping vector to array for transforming
Y_train_data = Y_train_data.reshape((len(Y_train_data), 1))
Y_train_data = sc.fit_transform(Y_train_data)
#print(Y_train_data)
print("--"*20)
#converting back to vector
Y_train_data = Y_train_data.ravel()
print(Y_train_data)
#print("--"*20)

X_test = sc.transform(X_test)
# Scaling Splitted training and val sets
X_train = sc.fit_transform(X_train)
X_val = sc.fit_transform(X_val)
print(X_train)
print("--"*20)
print(X_val)
print("--"*20)
#Reshaping vector to array for transforming
Y_train = Y_train.reshape((len(Y_train), 1))
Y_train = sc.fit_transform(Y_train)
print(Y_train)
print("--"*20)
#converting back to vector
Y_train = Y_train.ravel()
Y_train
```

```python
def score(y_pred, y_true):
   error = np.square(np.log10(y_pred +1) - np.log10(y_true +1)).mean() ** 0.5
   score = 1 - error
   return score,error
y_true = Y_val
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
Y_pred = sc.inverse_transform(lr.predict(X_val))
for i in range(len(Y_pred)):
   if Y_pred[i] < 0:
       Y_pred[i] = 0

print("\n Linear Regression SCORE  and error: ", score(Y_pred, y_true))

mlab.scatter(y_true, Y_pred,  color='#f57e42')
mlab.plot(y_true,y_true)
```

```python
#Calculate model metrics
from sklearn import metrics
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_true, Y_pred))  
print('Mean Squared Error:', 
      metrics.mean_squared_error(y_true, Y_pred))  
print('Root Mean Squared Error:', 
      np.sqrt(metrics.mean_squared_error(y_true, Y_pred)))

# Calculate R squared
# Goodness of Fit or Coefficient of Determination
r_squared = metrics.r2_score(y_true, Y_pred)

print(r_squared)
```
